```python
df_root = '/Volumes/dslab/2-stage-PGPIG/eval_results/'
step_92 = 'nted_fullstep_3_92.csv'
step_85 = 'nted_fullstep_3_85.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os 

df_92 = pd.read_csv(os.path.join(df_root, step_92))
df_85 = pd.read_csv(os.path.join(df_root, step_85))
df_92 = df_92.set_index(df_92.columns[0])
df_85 = df_85.set_index(df_85.columns[0])
```

```python
def draw_graph(df1, df2): 
    num_metrics = len(df1.index)
    fig, axes = plt.subplots(num_metrics, 1, figsize=(15, 6 * num_metrics))

    for i, metric in enumerate(df1.index):
        axes[i].plot(df1.columns, df1.loc[metric], label='DataFrame 1')
        axes[i].plot(df2.columns, df2.loc[metric], label='DataFrame 2')
        axes[i].set_title(f'Step-wise Evaluation Scores for {metric}')
        axes[i].set_xlabel('Steps')
        axes[i].set_ylabel('Evaluation Score')
        axes[i].legend()

    plt.tight_layout()
    plt.show()
draw_graph(df_92, df_85)
```
